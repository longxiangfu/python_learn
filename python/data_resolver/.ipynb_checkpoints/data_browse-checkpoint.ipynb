```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import os

dirpath = 'E:\\pycharmWork\\python\\data_resolver'
data = pd.read_csv(os.path.join(dirpath, 'athlete_events.csv'))
region = pd.read_csv(os.path.join(dirpath, 'noc_regions.csv'))
```

```python
# 前5行
head5 = data.head(5)
print(head5)
# 生成描述信息，数据分布的集中趋势，离散度
des = data.describe()
print(des)
```

```python
# data的信息
data.info()
```

```python
region_head5 = region.head(5)
print(region_head5)
```

```python
# 将data和region进行左连接，同mysql的左连接
merged = pd.merge(data, region, on='NOC', how='left')
merged_head5 = merged.head(5)
print(merged_head5)
```

金牌获得者年龄分布

```python
# 过滤出金牌数据
goldMedals = merged[(merged.Medal == 'Gold')]
goldMedals_head = goldMedals.head()
print(goldMedals_head)
```

```python
# 绘制一个金牌获得者年龄分布图，查看数据分布
# 检查年龄是否有NaN
goldMedals.isnull().any()
```

```python
# 使用年龄为非NaN的数据
goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
print(goldMedals.head())
```

```python
# 绘制
plt.figure(figsize=(20, 10))  # 画布大小
# plt.tight_layout()  # 紧凑型布局 因为会报错所以注释掉了
sns.countplot(goldMedals['Age'])  # 绘制图形    x轴为Age, y轴为对应的数据的数量

from matplotlib.pylab import mpl
from matplotlib.font_manager import FontProperties

fontPath = 'E:\\pycharmWork\\python\\data_resolver\\毛笔书法字体(启功体)繁启体.TTF'
my_font = FontProperties(fname=fontPath)
mpl.rcParams['axes.unicode_minus'] = False
plt.title('金牌获得者年龄分布', fontproperties=my_font)  # 图形title
plt.show()  # 显示图形
```

```python
# 金牌获得者，年龄大于50，从事的运动项目分布
# 查看符合条件的数据量
goldMedals[goldMedals['Age'] > 50]['ID'].count()
```

```python
# 绘制
older50 = goldMedals[goldMedals['Age'] > 50]
plt.figure(figsize=(20, 10))  # 画布大小
sns.countplot(older50['Sport'])  # 绘制图形    x轴为Age, y轴为对应的数据的数量

from matplotlib.pylab import mpl
from matplotlib.font_manager import FontProperties

fontPath = 'E:\\pycharmWork\\python\\data_resolver\\毛笔书法字体(启功体)繁启体.TTF'
my_font = FontProperties(fname=fontPath)
mpl.rcParams['axes.unicode_minus'] = False
plt.title('金牌获得者年龄大于50岁的运动项目分布', fontproperties=my_font)  # 图形title
plt.show()  # 显示图形
```

女子夏季运动数据分析

```python
# 过滤
woman_data = merged[(merged.Sex == 'F') & (merged.Season == 'Summer')]
print(woman_data.head())
```

```python
# 绘制  x轴：年份    y轴：对应的数据量即夏季运动会女子运动员奖牌数量
plt.figure(figsize=(20, 10))
# sns.countplot(x='Year', data=woman_data)
sns.countplot(woman_data['Year'])
plt.title('每届夏季运动会女子奖牌数据分布', fontproperties=my_font)
plt.show()
# woman_data['ID'][woman_data['Year'] == 1900].count()
woman_data['ID'].loc[woman_data['Year'] == 1900].count()
```

奥运会随时间的变化

1、男女数量的变化  2、运动员年龄、体重、身高的变化

```python
# 1、男女运动数量随时间的变化（夏季运动会）
MenOverTime = merged[(merged['Sex'] == 'M') & (merged['Season'] == 'Summer')]
WoMenOverTime = merged[(merged['Sex'] == 'F') & (merged['Season'] == 'Summer')]
```

```python
MenOverTime.head()
```

```python
part = MenOverTime.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 8))
part.plot()
# part.loc[:, 'M'].plot()  # 折线图
plt.title('男性运动员人数随时间的变化', fontproperties=my_font)
```

```python
part = WoMenOverTime.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 8))
# part.plot() # 折线图  x轴为year sex    y轴为数量
part.loc[:, 'F'].plot()  # 折线图
plt.title('女性运动员人数随时间的变化', fontproperties=my_font)
```

```python
MenOverTime.loc[MenOverTime['Age'] > 80].head(10)
```

```python
# 2、运动员年龄随时间的变化
plt.figure(figsize=(20, 10))
sns.boxplot('Year', 'Age', data=MenOverTime)  # 箱型图（盒须图） x为Year y为Age,图中黑色的点为异常值
plt.title('男性运动员年龄随时间的变化', fontproperties=my_font)
```

```python
# 3、运动员体重随时间的变化
plt.figure(figsize=(20, 10))
sns.pointplot('Year', 'Weight', data=MenOverTime)# 点状图
plt.title('男性运动员体重随时间的变化', fontproperties=my_font)
```

```python
plt.figure(figsize=(20, 10))
sns.pointplot('Year', 'Weight', data=WoMenOverTime)
plt.title('女性运动员体重随时间的变化', fontproperties=my_font)
```
